```python
# Import libraries
import glob
import numpy as np
import matplotlib.pyplot as plt
import corner
from IPython.display import display, Math
plt.rcParams['font.size'] = '17'
```

```python
x=np.arange(0,1.01,0.01)
y=np.arange(0,1.01,0.01)
xi, yi = np.meshgrid(x, y)
rho = (xi-yi)
rho[rho<0] = 0
apop = (xi-yi)
for i in range(len(x)):
    for j in range(len(y)):
        apop[i][j] = 1+(1.0/(1+np.exp(-100*(y[j]-x[i]))))
        apop[i][j] = 1.0-np.exp(-1.0*apop[i][j])
        rho[i][j] = max(x[i]-y[j],0)
        if y[j] < 1:
            rho[i][j] = rho[i][j]/(1-y[j])
            rho[i][j] = 1.0-np.exp(-1.0*rho[i][j])
fig, axes = plt.subplots(1,2, figsize=(15, 5))

ax1 = axes[0]
ax1.set_ylabel('$\sigma$')
ax1.set_xlabel('$\sigma_H$');
#im1 = ax1.imshow(apop, cmap='jet', vmin=apop.min(), vmax=apop.max(), origin='lower',extent=[0, 1, 0, 1])
im1 = ax1.imshow(apop, cmap='jet', vmin=0, vmax=1, origin='lower',extent=[0, 1, 0, 1])
fig.colorbar(im1, ax=ax1,label='P(D|Q)')

ax2 = axes[1]
ax2.set_ylabel('$\sigma$')
ax2.set_xlabel('$\sigma_H$');
#im2 = ax2.imshow(rho, cmap='jet', vmin=rho.min(), vmax=rho.max(), origin='lower',extent=[0, 1, 0, 1])
im2 = ax2.imshow(rho, cmap='jet', vmin=0, vmax=1, origin='lower',extent=[0, 1, 0, 1])
#im2 = ax2.imshow(rho, cmap='jet', vmin=rho.min(), vmax=rho.max(), origin='lower',extent=[xi.min(), xi.max(), yi.min(), yi.max()])
fig.colorbar(im2, ax=ax2,label='P(P|Q)')
plt.savefig('stochastic_processes_same.pdf', bbox_inches = 'tight', pad_inches = 0.02)
plt.show()
```
